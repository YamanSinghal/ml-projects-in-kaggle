# For converting all reviews into lower case for stability

```python
#function to convert to lower case
def lower_case(text):
    return text.lower()
```

```python
data["review"] = data["review"].apply(lower_case)
```

```python
data["review"][0]
```

# For removing all html tags from all reviews

```python
#for deleting html tags
```

```python
import re
clean = re.compile("<.*?>")
re.sub(clean," ",data.iloc[2543].review)
```

```python
def remove_tags(text):
    clean = re.compile("<.*?>")
    return re.sub(clean," ",text)

```

```python
data["review"] = data["review"].apply(remove_tags)
```

```python
data["review"][0]
```

# For removing all special characters 

```python
#to remove special characters
```

```python
def remove_special(text):
    x = ""
    for i in text:
        if i.isalnum():
            x = x + i
        else:
            x = x + " "
    return x
        
```

```python
remove_special("hi @yaman")
```

```python
data["review"] = data["review"].apply(remove_special)
```

# For removing stop words

```python
import nltk
```

```python
df= data.sample(5000)
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words("english")   #all stopping words in english language
```

```python
def remove_stop(text):
    x = []
    for i in text.split():
        if i not in stopwords.words("english"):
            x.append(i)
    y = x[:]
    x.clear()
    return y
```

```python
remove_stop("yaman and aman")
```

```python
df["review"] = df["review"].apply(remove_stop)
```

```python
df.head()
```

```python
# For stemming
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
y = []
def stemming(text):
    for i in text:
        y.append(ps.stem(i))
    z = y[:]
    y.clear()
    return z
```

```python
stemming(["love","loved","lovely"])
```

```python
df["review"] = df["review"].apply(stemming)
```

```python
df.head()
```

# Since the review are in list so to convert them back to string

```python
def join(text):
    return " ".join(text)
#the input here should be in list and since our review columns is already in list so no need to convert that
```

```python
df["review"] = df["review"].apply(join)
```

# Now for training model the data has to be converted in a format so that the columns are all the words in reviews after pre processing and the rows are all the reviews. the values in them are the frequency of each word written in each review

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
X = cv.fit_transform(df["review"]).toarray()
#it returns a sparse matrix
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 89)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
clf1 = GaussianNB()
clf2 = MultinomialNB()
clf3 = BernoulliNB()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
clf3.fit(X_train,y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
y_pred3 = clf3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred1,y_test))
print(accuracy_score(y_pred2,y_test))
print(accuracy_score(y_pred3,y_test))
```
